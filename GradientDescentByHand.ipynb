```python
def f(x):
    return x**2

# Range
x = np.linspace(-10, 10, 500)
y = f(x)

# Specific point to highlight
x_point = 3
y_point = f(x_point)

# Plot curve
plt.plot(x, y, label='f(x) = x²')

# Highlight point
plt.scatter(x_point, y_point, color='red', zorder=5)
plt.annotate(f'({x_point}, {y_point})',
             (x_point, y_point),
             textcoords="offset points",
             xytext=(10,10),
             ha='center',
             color='red')

# Decorations
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x² with a Highlighted Point')
plt.legend()
plt.show()
```
